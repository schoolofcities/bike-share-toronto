```python
import pandas as pd
import os

```

```python
#read all the csv and merge into one large dataframe for processing. 
ridership_full = pd.read_csv(f"bikeshare-ridership-2023/Bike share ridership 2023-01.csv", encoding= 'cp1252')
for file in os.listdir("bikeshare-ridership-2023")[1:]:
    csv = pd.read_csv(f"bikeshare-ridership-2023/{file}", encoding= 'cp1252')
    ridership_full = ridership_full.append(csv)
    
ridership_full   
```

```python
ridership_clean = ridership_full.loc[~((ridership_full['Start Station Id'] == ridership_full['End Station Id']) & (ridership_full['Trip  Duration'] < 300))]
#ridership_clean = ridership_clean[['Trip  Duration', 'Start Station Id', 'End Station Id']]
#remove the decimal point
ridership_clean['End Station Id'] = ridership_clean['End Station Id'].fillna(0).astype('int')
#merge start and end station id to identify unique value counts
ridership_clean['Start End Station'] = ridership_clean['Start Station Id'].astype('str') + ", " + ridership_clean['End Station Id'].astype('str')
ridership_clean
```

```python

#read the bikeshare stations sheet with coordinates of the stations
bikeshare_stations = pd.read_csv("bikeshare station id.csv",encoding= 'cp1252')
bikeshare_stations['ID'] = bikeshare_stations['ID'].astype('str')
bikeshare_stations
```

```python
ridership_clean['Start Station Id'] = ridership_clean['Start Station Id'].astype('str')

bikeshare = pd.merge(ridership_clean, bikeshare_stations, left_on = "Start Station Id", right_on = "ID")

bikeshare
```

```python
#calculating each number of trips that each station is a destination. 
bikeshare_ending = pd.pivot_table(bikeshare, index = 'Start Station Id', columns = 'End Station Id', values = 'Station', aggfunc = 'count').reset_index()
#bikeshare_ending = bikeshare_ending.drop(0, axis =1)
bikeshare_ending
#the column names are the end station Id, and the values under Start Station ID are the ID of starting station. When display on web, for instance the selected column is 7000, 
#each value is the number of trips ending at station 7000
```

```python
bikeshare_ending['Start Station Id'] = bikeshare_ending['Start Station Id'].astype('str')
ridership_ending = pd.merge(bikeshare_ending, bikeshare_stations, left_on = "Start Station Id", right_on = "ID")
ridership_ending = ridership_ending.drop(['Name', 'ID'], axis =1)
ridership_ending.to_csv("bikeshare relations (destination).csv")
```

```python
ridership_ending
```

```python
#calculating each number of trips that each station is a origin. 
bikeshare_starting = pd.pivot_table(bikeshare, index = 'End Station Id', columns = 'Start Station Id', values = 'Station', aggfunc = 'count').reset_index()
#bikeshare_starting = bikeshare_starting.drop(0)
bikeshare_starting
#the column names are the end station Id, and the values under Start Station ID are the ID of starting station. When display on web, for instance the selected column is 7000, 
#each value is the number of trips ending at station 7000
```

```python
bikeshare_starting['End Station Id'] = bikeshare_starting['End Station Id'].astype('str')
ridership_starting = pd.merge(bikeshare_starting, bikeshare_stations, left_on = "End Station Id", right_on = "ID")
ridership_starting = ridership_starting.drop(['Name', 'ID'], axis =1)
ridership_starting.to_csv("bikeshare relations (origin).csv")
```
